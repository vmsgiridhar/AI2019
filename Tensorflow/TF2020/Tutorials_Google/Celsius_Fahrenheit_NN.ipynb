# Predicting Fahrenheit temperature for the given Celsius temperature using Linear Regression and Neural Network

# Abstract

## We all came across this forrmula in our Physics or Mathematics classes that converts a centigrade temperature into Fahrenheit temperature.

    Temperature in Fahrenheit = 9/5 * (Temperature in Celsius) + 32
    
                                   or
    Temperature in Fahrenheit = 1.8 * (Temperature in Celsius) + 32

## When we talk about Machine Learning and supervised learning in specific, supplied data of history (in our case Celsius and respective Fahrenheit temperatures), the Machine should be capable of learning the aforementioned formuala to calculate Fahrenheit temperature for a given Celsius temparature by itself.

## Now let us build the basic Linear Regression Model and also a Neural Network model for the same.

## Part 1 - Linear Regression Model

```python
# importing the required libraries
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# data for the ML algorithm
celsius = np.array([-67.0, -34.0, 0, 34.0, 54.0, 67.0, 100.0], dtype=float)
fahrenheit = np.array([-88.6, -29.2, 32, 93.2, 129.2, 152.6, 212.0], dtype=float)
# INFO: celsius is the feature and fahrenheit is the label we are trying to predict here.
```

```python
# Plotting the relation between celsius and fahrenheit - a perfect straight line
plt.plot(celsius, fahrenheit)
```

```python
# Partitioning the existing data into Train and Test sets
X_train, X_test, y_train, y_test = train_test_split(celsius, fahrenheit, test_size=0.30)
```

```python
# Fitting a simple linear Regressor onto the data
lin_reg = LinearRegression()

# reshaping the numpy arrays (1D) to 2D using reshape(-1,1)
X_train = X_train.reshape(-1,1)
y_train = y_train.reshape(-1,1)

# fit data to regressor
lin_reg.fit(X_train, y_train)
```

```python
# Let us now see how the lin_reg regressor predicted results on the X_test set
X_test = X_test.reshape(-1,1)
y_test = y_test.reshape(-1,1)
y_pred = lin_reg.predict(X_test)
y_pred
```

```python
# Let us see the y_test - You may observer y_pred is super close to the y_test
y_test
```

```python
# Getting the intercept and coefficient for this regressor
# this should be close to 32
print('Intercept of the Linear Regression Model is: {}'.format(lin_reg.intercept_[0]))

# similarly, calculating the coefficient
# this should be close to 1.8
print('Coefficient of the Linear Regression Model is: {}'.format(lin_reg.coef_[0]))

# Let us also calculate the MSE (Mean-squared error)
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
# If MSE is Mean Squared error, let us find the mean error
mse ** 0.5
```

```python
# Visualising the Training set results
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, lin_reg.predict(X_train), color = 'blue')
plt.title('Celsius vs Fahrenheit (Training set)')
plt.xlabel('Celsius')
plt.ylabel('Fahrenheit')
plt.show()
```

```python
# Visualising the Test set results
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_test, lin_reg.predict(X_test), color = 'blue')
plt.title('Celsius vs Fahrenheit (Test set)')
plt.xlabel('Celsius')
plt.ylabel('Fahrenheit')
plt.show()
```

```python
# Predicting a new value with the regressor with coefficient and intercept
lin_reg.predict([[-60]])[0][0]
```

### To verify if this is correct, we need to verify if the linear regressor generalized it to match the general formula used to convert a Centigrade temperature to Fahrenheit temperature.

```python
# generic formula for converting centigrade to fahrenheit temperature.

def centigrade_to_fahrenheit(cent):
    return (1.8 * cent + 32)
```

```python
# this regressor gives accurate answers.
lin_reg.score(X_test, y_test)
```

```python
centigrade_to_fahrenheit(50000)
```

```python
lin_reg.predict([[50000]])[0][0]
```

## Part 2 - Using a Neural Network

```python
# We have a very basic problem here, so a Neural Network with a single layer and a single neuron is enough
```

```python
# importing tensorflow
import tensorflow as tf
```

### Building the layer with a single neuron
* input_shape=[1] — This specifies that the input to this layer is a single value. That is, the shape is a one-dimensional array with one member. Since this is the first (and only) layer, that input shape is the input shape of the entire model. The single value is a floating point number, representing degrees Celsius.

* units=1 — This specifies the number of neurons in the layer. The number of neurons defines how many internal variables the layer has to try to learn how to solve the problem (more later). Since this is the final layer, it is also the size of the model's output — a single float value representing degrees Fahrenheit. (In a multi-layered network, the size and shape of the layer would need to match the input_shape of the next layer.)

```python
# building the layer
layer0 = tf.keras.layers.Dense(units=1, input_shape=[1])
```

```python
layer0.get_config()
```

```python
# We use sequential model and add the layer0 to it
model = tf.keras.Sequential([layer0])
```

```python
model.get_config()
```

### Before training, the model has to be compiled. When compiled for training, the model is given:

* Loss function — A way of measuring how far off predictions are from the desired outcome. (The measured difference is called the "loss".)

* Optimizer function — A way of adjusting internal values in order to reduce the loss.

```python
model.compile(loss='mean_squared_error',
              optimizer=tf.keras.optimizers.Adam(0.1))

# More about adam optimizer here - https://machinelearningmastery.com/adam-optimization-algorithm-for-deep-learning/
```

### Model training

```python
process = model.fit(X_train, y_train, epochs=2500, verbose=2)
```

### Plotting how the error got reduced eventually

```python
plt.xlabel('Epoch')
plt.ylabel('Loss Magnitude')
plt.plot(process.history['loss'])
```

```python
# process.history is a dictionary
process.history
```
